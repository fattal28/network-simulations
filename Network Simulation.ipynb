## Network

```python
from itertools import combinations
from networkx import nx
import matplotlib.pyplot as plt
import random

# probability for an edge to exist
p = 0.25

nodes = []
for i in range(25):
    nodes.append(i)

g = nx.Graph()
g.add_nodes_from(nodes)

for i, j in combinations(g, 2):
    if random.random() < p:
        g.add_edge(i, j)

nx.draw(g)
plt.show()
```
